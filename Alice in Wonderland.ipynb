# Finding related characters in Lewis Carrolls' "Alice in Wonderland"

The purpose of this notebook is to show how to use the Gensim library (http://radimrehurek.com/gensim/) to find related characters and keywords appearing in text. The workflow is inspired from Sujit Pal's blog (http://sujitpal.blogspot.cz/2014/08/topic-modeling-with-gensim-over-past.html). 

So let's begin by importing some libraries:

```python
from gensim.parsing.preprocessing import STOPWORDS
from sklearn.cluster import KMeans
import gensim
import numpy as np
import string
```

Now we need to parse the text, getting rid of punctuation marks, blank spaces and turning everything to lowercase. 


```python
def parse(line):
    return line.translate(string.maketrans("",""), string.punctuation).lower().split() 
```

```python
INPUT_FILE = './data/alice.txt'
```

We should filter out some words that are useless for our purposes, like prepositions or pronouns, called stopwords. 
Gensim has a list of those, but we'll throw some more in, which we obtained iteratively after looking at the results we obtained.

```python
stopwords = [u'project',u'gutenbergtm',u'looked',u'gutenberg',u'electronic',u'got',u'use',u'don\'t', u'him.',u'"you',u'"i'
             u'certain',u"don't", u'it.',u'my', u'work,',u'got', u'paragraph', u'work.', u'set',u'domain',u'public', 
             u'copyright',u'license',u'terms',u'contact', u'copyright',u'donations',u'alice,',u'archive',u'alice.',u'added'
             u"'i", u'said',u'foundation',u'work', u'little',u'went',u'it,', "'i", u'dont',u'works',u'wont',u'began',u'know',
             u'like',u'im',u'ive',u'oh', u'think',u'thing',u'thats',u'agreement',u'thought',u'added',u'ill',u'right',u'tell',
             u'youre',u'long',u'cook', u'mock',u'look', u'going', u'literary', u'asked', u'soon',u'voice', u'small', 
             u'minute', u'come', u'states',u'question',u'gave',u'dance',u'course',u'march',u'dear',u'shall',u'poor',
             u'head',u'came', u'sat', u'sort', u'way',u'end', u'looking', u'getting', u'things', u'away', u'round', 
             u'theyre',u'cried',u'replied',u'copy',u'half',u'change',u'moment',u'wish', u'begin',u'left',u'time', 
             u'alice',u'doesnt',u'felt',u'knew',u'day',u'tone',u'till',u'hare',u'theres',u'hand',u'hastily',u'hear', 
             u'seen',u'sure',u'turned',u'beginning',u'tried',u'didnt',u'near',u'glad',u'should',u'makes',u'repeated',
             u'kept',u'large',u'heard',u'talking',u'spoke', u'trademark', u'united',u'refund', u'thinking', u'access',
             u'fee', u'ran',u'written',u'let',u'silence',u'great', u'join',u'laws']
for word in STOPWORDS:
    stopwords.append(word)
```

Now let's create a dictionary. This is simply the vocabulary from the book, adding an integer id to each unique word. We will remove words appearing only once, as well as words appearing more than a certain threshold, which is set after getting some feedback.

```python
dictionary = gensim.corpora.Dictionary(parse(line) for line in open(INPUT_FILE))
THRESHOLD = 0.6
stop_ids = [dictionary.token2id[stopword] for stopword in stopwords if stopword in dictionary.token2id]
once_ids = [tokenid for tokenid,docfreq in dictionary.dfs.iteritems() if docfreq ==1]
freq_ids = [tokenid for tokenid,docfreq in dictionary.dfs.iteritems() if docfreq >= THRESHOLD*len(dictionary)]
dictionary.filter_tokens(stop_ids+once_ids)
dictionary.compactify()
```

Once the dictionary is set, we create the corpus. The corpus is a collection of tuples (id,score), where id is the integer id of the word created on the dictionary in the previous step, and score is the frequency of the word. 

```python
corpus = [dictionary.doc2bow(parse(line)) for line in open(INPUT_FILE)]
```

Now we create the tfidf representation of the corpus. The tf-idf score of each word tells us how important that word is on the given line with respect to the rest of the book.


```python
tfidf = gensim.models.TfidfModel(corpus, normalize=True)
corpus_tfidf = tfidf[corpus]
```

So what's going on? We represent the book as a huge (but sparse) matrix where each row is a line of the book and the columns are the words (the dictionary), and on each non-zero entry we write down the tf-idf score. 

What we want is to decompose this matrix in  a smaller subspace, reducing collinear rows. This is achieved by singular values (LSI, latent semantic indexing).

We have no idea of how many classes of characters-context (topics) should there be. But we can look at the two dimensional projection, and look at the total inertia as a function of the number of clusters. The inertia is the sum of square distances from every point to its nearest centroid.

```python
max_k = 10
ks = range(1,max_k+1)

lsi = gensim.models.LsiModel(corpus_tfidf,id2word=dictionary, num_topics = 2)
X=[]
for vector in lsi[corpus]:
    if len(vector)!= 2:
        continue
    X.append(np.array([vector[0][1],vector[1][1]]))
X = np.array(X)

# Calculate the third derivative of the inertia

inertias = np.zeros(max_k)
diff = np.zeros(max_k)
diff2 = np.zeros(max_k)
diff3 = np.zeros(max_k)
for k in ks:
    kmeans = KMeans(k).fit(X)
    inertias[k - 1] = kmeans.inertia_
    # first difference    
    if k > 1:
        diff[k - 1] = inertias[k - 1] - inertias[k - 2]
    # second difference
    if k > 2:
        diff2[k - 1] = diff[k - 1] - diff[k - 2]
    # third difference
    if k > 3:
        diff3[k - 1] = diff2[k - 1] - diff2[k - 2]

elbow = np.argmin(diff3[3:]) + 3
```

We compute the third "derivative", to get the inflection points. The edge of the elbow is then the minimum.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(ks,inertias)
plt.plot(ks[elbow],inertias[elbow], marker='o')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.show()
```

So this suggests that the number of clusters/topics is 5. Let's take a closer look at the clusters, to see if this is reasonable. We may not get anything very spectacular in general, since we can plot only a projection, but it is worth trying for sanity check.

```python
kmeans = KMeans(elbow).fit(X)
y = kmeans.labels_
colors = ['b','g','r','m','c']

for i in range(X.shape[0]):
    plt.scatter(X[i][0],X[i][1], c = colors[y[i]], s = 10)
plt.show()    
```

Finally, we run Latent Dirichlet Allocation. This model assumes that each line in the book is generated as a mixture of topics, from which the number is known, and tries to get back the topics that generated them. In our case, a topic corresponds to certain contexts on which the characters appeared, or certain events happened (for instance, the tea party, the trial, etc.) So the topics are the "main events" of the book, somehow.

```python
NUM_TOPICS = ks[elbow]

lda = gensim.models.LdaModel(corpus, id2word=dictionary, num_topics=NUM_TOPICS, passes = 10)
lda.print_topics(NUM_TOPICS,num_words=20)
```

Finally, some word cloud visualizations to better interprete the results, using Andreas Mueller package: (https://github.com/amueller/word_cloud)

```python
import wordcloud
import matplotlib.pyplot as plt
%matplotlib inline

final_topics = lda.print_topics(NUM_TOPICS)
curr_topic = 0
for line in final_topics:
    #line = line.strip()[line.rindex("") + 2:]
    scores = [float(x.split("*")[0]) for x in line.split(" + ")]
    words = [x.split("*")[1] for x in line.split(" + ")]
    freqs = []
    wc = wordcloud.WordCloud()
    for word, score in zip(words, scores):
        freqs.append((word, score))
    elements = wc.fit_words(freqs)
    plt.imshow(wc)
    plt.show()
```

Sounds credible? It sure requires a lot of tweaking, but the results are not terrible :).
